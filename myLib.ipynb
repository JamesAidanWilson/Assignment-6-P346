```python
# Factorial of n
def factorial(n):
    n = float(n)
    if n >= 1 and float(n)==int(n):
        n = int(n)
        fact = 1
        for i in range(1,n+1):
            fact = fact*i
        return fact
    elif int(n) == 0:
        return 1
    else:
        print("Invalid Input!")
        return None
```

```python
# READING A MATRIX FROM A .txt FILE

def readMat(file_name):
    file=open(str(file_name),"r")
    M=[[]]
    temp=''
    i=0
 #what follows below is a just accessing the strings one by one, and and converting the file data to list
 #of lists format
    for j in file:
        for k in j:
            if k=='\n':
                M[i]+=[float(temp)]
                temp=''
                i+=1
                M+=[[]]
            elif k!=',' and k!='\n':
                temp+=k
            elif k==',' or k=='':
                M[i]+=[float(temp)]
                temp=''
    return(M)

#Function for creating matrix.
def newMat(dat,r,c):
        mat = []
        for i in range(r):
            temp = []
            for j in range(c):
                temp.append(0)
            mat.append(temp)         #Creating a null matrix.
        for i in range(r):
            for j in range(c):
                    mat[i][j] = float(dat[c*i+j])
        return mat

# Function for printing a matrix in proper format.
def printMat(mat):
    r = len(mat)
    c = len(mat[0])
    print("[",end="")
    for i in range(r):
        for j in range(c):
            temp = str(mat[i][j])
            if i!=0 and j == 0:
                print(" "+temp+"     ",end="")   
            elif j==c-1:
                print(temp.rstrip(),end="")
            elif j!=c-1:
                print(temp+"     ",end="")
        if i!= r-1:
            print("\n")
        else:
            print("]")
    print()

    
# Multiplication of matrices
def prodMat(A,B): 
    r1 = len(A)
    r2 = len(B)
    c1 = len(A[0])
    c2 = len(B[0])
    if c1 == r2:              #Checking eligibility for multiplication.
        ans = []
        for i in range(r1):
            temp = []
            for j in range(c2):
                temp.append('0')
            ans.append(temp)         #Creating a null answer matrix.
        for i in range(r1):
            for j in range(c2):
                temp = 0
                for k in range(c1):
                    temp = temp + float(A[i][k])*float(B[k][j])
                ans[i][j] = temp
        return ans
    else:
        print("Invalid input!")
        return None
 
 # Create a null matrix of order n.
def create0(r,c):
    dat = "0"*r*c
    I = newMat(dat,r,c)
    return I
    

# Create Identity matrix of order n.
def createI(n):
    dat = "0"*n**2
    I = newMat(dat,n,n)
    i=0
    for x in I:
        x[i] = 1
        i+=1
    return I

# Extracting leading submatrices
def leadingsubMat(A,n):    # n = No. of the submatrix, e.g. for [[0 1 2][3 4 5][6 7 8]], the 1st leading submatrix
    r = len(A)
    c = len(A[0])
    if n>r or n>c:
        print("Dimension of submatrix exceeds that of actual matrix")
        return None
    else:
        sub=newMat(("0"*n*n),n,n)  # reads [[0]], 2nd one reads [[0 1][3 4]] and the 3rd one reads [[0 1 2][3 4 5][6 7 8]].
        for i in range(n):
            for j in range(n):
                sub[i][j] = A[i][j]
        return sub

# Extracting a column from a given matrix and return it as a column vector
def extractCol(A,m):         # Extracts the mth column of A
    r = len(A)
    c = len(A[0])
    if m>r or m>c:
        print("Dimension of submatrix exceeds that of actual matrix")
        return None
    else:
        col = []
        for i in range(r):
            temp=[]
            temp.append(A[i][m-1])
            col.append(temp)
        return col    

# Extracting a row from a given matrix and return it as a column vector
def extractRow(A,m):         # Extracts the mth row of A
    r = len(A)
    c = len(A[0])
    if m>r or m>c:
        print("Dimension of submatrix exceeds that of actual matrix")
        return None
    else:
        col = create0(r,1)
        for i in range(r):
            col[i][0] = A[m-1][i]
        return col    
    
# R = k*R
def scalerow(R,k,n):
    for i in range(n):
        R[i] = k*float(R[i])
    return R

# R1 = R1 - R2
def deductrow(R1,R2,n):
    for i in range(n):
        R1[i] = R1[i] - R2[i]
    return R1

# Finding the maximum value of a column
def maxMat(A,i1,i2,c): # Checks between A(i1)(c) and A(i2)(c)
    max = 0
    for i in range(i1,i2+1):
        if A[i][c] > max:
            max = A[i][c]
    return max

# Swap two rows
def rowswap(A,i1,i2):
    temp = A[i1]
    A[i1] = A[i2]
    A[i2] = temp
    return A

# Partial Pivoting for Gauss-Jordan Elimination
def partialpivot(Ab,m):  # Ab is the Augmented matrix; m is the pivot index
    r = len(Ab)
    c = len(Ab[0])
    swaps = 0
    pivot = Ab[m][m]
    max = maxMat(Ab,m,r-1,m)
    if pivot!=0 and pivot==max:
        return Ab, swaps
    if m==(r-1):
        if pivot==0:
            Ab = None
            swaps = 0
        return Ab, swaps
    else:
        for j in range(m+1,r):
            if Ab[j][m]!=0 and Ab[j][m]==max:
                Ab = rowswap(Ab,m,j)
                swaps+=1
                pivot = Ab[m][m]        
        if pivot==0:
            swaps=0
            Ab=None
        return Ab, swaps
        
# Gauss-Jordan Elimination
def gaussj(Ab,str=None): # For checking the determinant user may type 'det' in place of str, or 'inv' for inverse.
    r = len(Ab)
    c = len(Ab[0])
    flag = 0
    det = 1
    for i in range(r):
        Ab, swaps = partialpivot(Ab,i)
        flag+=swaps
        if Ab is None and str is None:
            return("No unique solution exists for the given system.")
            exit()
        elif Ab is None and str=='det':
            return("Indeterminate value found/Partial pivoting failed.")
            exit()
        elif Ab is None and str=='inv':
            return("Indeterminate value found/Partial pivoting failed.")
            exit()
        else:    
            det = det*Ab[i][i]            # Stores the determinant
            ratio=1/Ab[i][i]
            Ab[i] = scalerow(Ab[i],ratio,c)    # R[i] = R[i]/pivot
            for j in range(0,r):
                if i!=j and Ab[j][i]!=0:      # element is already ZERO.
                    factor = Ab[j][i]
                    for k in range(i,c):
                        Ab[j][k]-= factor * Ab[i][k]
    
    Ab = roundMat(Ab,2)
    
    if str==None:
        print("The solns are:")
        print()
        for i in range(4):    
            print(Ab[i][4])
            print()
    elif str=='inv':
        return Ab
    elif str=='det':
        if det!=None:
            return det*(-1)**flag
        else:
            return det

# Determinant of a matrix (Using Gauss Jordan El.)
def detMat(A):
    r = len(A)
    c = len(A[0])
    if r==c:
        det = gaussj(A,'det')
        return det
    else:
        print("Enter a square matrix")
        return None

# Inverse of matrix using Gauss-Jordan El.
def InvMat(A):
    n = len(A)
    I = createI(n)
    b = 2*n
    M = newMat("0"*n*b,n,b)
    for i in range(n):
        for j in range(b):
            if j<n:
                M[i][j] = A[i][j]
            if j>=n:
                M[i][j] = I[i][j-n]
    M = gaussj(M,'inv')
    if M!="Indeterminate value found/Partial pivoting failed.":
        ans = createI(n)
        for i in range(n):
            for j in range(n,2*n):
                ans[i][j-n] = M[i][j]
        return ans
    else:
        return("Inverse doesn't exist/Partial pivoting failed.")

# For accepting a square matrix of order n from the user
def acceptMat(r,c):
    dat = "0"*r*c
    A = newMat(dat,r,c)
    for i in range(r):
        for j in range(c):
            A[i][j] = float(input(f"Enter A[{i}][{j}]: "))
    return A

# Rounding off the elements of a matrix correct upto 2 decimal places
def roundMat(A,k):
    r = len(A)
    c = len(A[0])
    # A, r, c = Matrix and its dimensions
    # k = no. of decimal places to round off to.
    for i in range(r):
        for j in range(c):
            A[i][j]=float(format(A[i][j],f'.{k}f'))
    return A

# Pivotizing a Matrix
def pivotize(A):
    r = len(A)
    c = len(A[0])
    for x in range(r):
        A,s = partialpivot(A,x)
    return A


# LU decomposition using Doolittle algorithm
def doolittle(A):
    r = len(A)
    c = len(A[0])  
    for i in range(r):
        #This section will check if the decompostion is possible
        #And if required will use partial pivoting to attempt factorisation
        sub = leadingsubMat(A,i+1)    # Exctracting the ith leading submatrix
        if detMat(sub)=="Indeterminate value found/Partial pivoting failed." or detMat(sub)==0: # detMat(sub) returns the 
            pP,s = partialpivot(A,i)                                                            # Determinant of submatrix
            if pP==None:
                return("LU Decomposition not possible")
                exit()
         
    #This section caclulates the value of the elements in L and U       
    for i in range(1,r):                 #choosing row number
        for j in range(c):               #choosing column number
            if i<=j:                     #Calculating u_{ij}
                temp=0
                for k in range(i):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=A[i][j]-temp     #replacing the original matrix with the calculated value
            elif i>j:                    #Calculating l_{ij}
                temp=0
                for k in range(j):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=(A[i][j]-temp)/A[j][j]
    A = roundMat(A,3)
    return A


# LU decomposition using Crout's algorithm
def crout(A):
    r = len(A)
    c = len(A[0])
    for i in range(r):
        #This section will check if the decompostion is possible
        #And if required will use partial pivoting to attempt factorisation
        sub = leadingsubMat(A,i+1)    # Exctracting the ith leading submatrix
        if detMat(sub)=="Indeterminate value found/Partial pivoting failed.":
            return("LU decomposition not possible")
            exit()
        if detMat(sub)==0:    # gaussj(sub,i,j,'det') returns the Determinant of the submatrix
            pP,s = partialpivot(A,i) 
            if pP==None:
                return("LU Decomposition not possible")
                exit()
    #This section caclulates the value of the elements in L and U       
    for j in range(1,r):                 #choosing row number
        for i in range(c):               #choosing column number
            if i>=j:                     #Calculating u_{ij}
                temp=0
                for k in range(j):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=A[i][j]-temp     #replacing the original matrix with the calculated value
            elif i<j:                    #Calculating l_{ij}
                temp=0
                for k in range(i):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=(A[i][j]-temp)/A[i][i]
    A = roundMat(A,3)
    return A


# LU decomposition using Cholesky's method
def cholesky(A):
    import math            # Importing math lib only to check square root
    r = len(A)            
    c = len(A[0])         
    for k in range(r):
        #This section will check if the decompostion is possible
        #And if required, will use partial pivoting to attempt factorisation
        sub = leadingsubMat(A,k+1)
        if detMat(sub)=="Indeterminate value found/Partial pivoting failed.":
            return("LU decomposition not possible")
            exit()
        if detMat(sub)==0:    # gaussj(sub,i,j,'det') returns the Determinant of the submatrix
            pP,s = partialpivot(A,i) 
            if pP==None:
                return("LU Decomposition not possible")
                exit()
    #This section caclulates the value of the elements in L and U
    #we dont need to calculate every combination of i,j since the matrix is symmetric
    for i in range(r):
        for j in range(i,r):
            if i==j:                   #diagonal elements
                temp=0
                for k in range(i):
                    temp+=A[i][k]**2
                A[i][i]=math.sqrt(A[i][i]-temp)
            if i<j:                    #calculating off diagonal elements
                temp=0
                for k in range(i):
                    temp+=A[i][k]*A[k][j]
                A[i][j]=(A[i][j]-temp)/A[i][i]
                A[j][i]=A[i][j]         #due to symmetricity of matrix
    A = roundMat(A,3)
    return A

# Forward substitution for LU 
def forsubs(L,b):
    r = len(L)
    c = len(b[0])
    X = create0(r,c)
    for i in range(r):
        for k in range(c):
            s = 0
            for j in range(i):
                s += L[i][j]*X[j][k]
            X[i][k] = (b[i][k]-s)*(1/L[i][i])
    return X

# Backward substitution for LU 
def backsubs(U,b):
    r = len(U)
    c = len(b[0])
    X = create0(r,c)
    for i in range(r-1,-1,-1):
        for k in range(c):
            s = 0
            for j in range(r-1,i,-1):
                s += U[i][j]*X[j][k]
            X[i][k] = (b[i][k]-s)*(1/U[i][i])
    return X

# Solving system of equations using Doolittle's algorithm
def solveDoolittle(Ab):
    r = len(Ab)
    c = len(Ab[0])
    
    # Separating A and b from the Augmented matrix
    A = create0(r,r)
    b = create0(r,1)
    for i in range(r):
        b[i][0] = Ab[i][-1]
        for j in range(c-1):
            A[i][j] = Ab[i][j]   # A and b are now separated
    
    A2 = A          # Creating a duplicate of A
    
    A2 = doolittle(A2)    # Performing LU decomposition on A using doolittle()
    
    # Separating L and U
    if A2!=None:          # Decomposing A2 into L and U using Doolittle
        L = create0(r,r)
        U = create0(r,r)
        for i in range(r):
            for j in range(r):
                if j<i:
                    L[i][j] = A2[i][j]
                if j>=i:
                    U[i][j] = A2[i][j]    # Separating L and U from the L/U augmented form obtained from doolittle()
                    if j==i:
                        L[i][i] = 1       # Making diagonal elements of L to 1
        
        y = forsubs(L,b)        # Solving Ly=b
        x = backsubs(U,y)       # Finally solving Ux=y
        x = roundMat(x,3)       # Using custom round function to round the matrix to 3 decimals
        return(x)
    else:
        return("No unique soln exists/LU Decomposition failed")
        exit()

    
# Solving system of equations using Crout's algorithm
def solveCrout(Ab):
    r = len(Ab)
    c = len(Ab[0])
    
    # Separating A and b from the Augmented matrix
    A = create0(r,r)
    b = create0(r,1)
    for i in range(r):
        b[i][0] = Ab[i][-1]
        for j in range(c-1):
            A[i][j] = Ab[i][j]   # A and b are now separated
    
    A2 = A          # Creating a duplicate of A
    
    A2 = crout(A2)    # Performing LU decomposition on A using crout()
    
    # Separating L and U
    if A2!=None:          # Decomposing A2 into L and U using crout
        L = create0(r,r)
        U = create0(r,r)
        for i in range(r):
            for j in range(r):
                if j<=i:
                    L[i][j] = A2[i][j]
                    if j==i:
                        U[i][i] = 1       # Making diagonal elements of U to 1
                if j>i:
                    U[i][j] = A2[i][j]    # Separating L and U from the L/U augmented form obtained from crout()
                            
        y = forsubs(L,b)        # Solving Ly=b
        x = backsubs(U,y)       # Finally solving Ux=y
        x = roundMat(x,3)
        return(x)
    else:
        return("No unique soln exists/LU Decomposition failed")
        exit()


# Solving system of equations using Cholesky's algorithm
def solveCholesky(Ab):
    r = len(Ab)
    c = len(Ab[0])
    
    # Separating A and b from the Augmented matrix
    A = create0(r,r)
    b = create0(r,1)
    for i in range(r):
        b[i][0] = Ab[i][-1]
        for j in range(c-1):
            A[i][j] = Ab[i][j]   # A and b are now separated
    
    A2 = A          # Creating a duplicate of A
    
    A2 = cholesky(A2)    # Performing LU decomposition on A using crout()
    
    # Separating L and U
    if A2!=None:          # Decomposing A2 into L and U using crout
        L = create0(r,r)
        U = create0(r,r)
        for i in range(r):
            for j in range(r):
                if j<=i:
                    L[i][j] = A2[i][j]
                    if j==i:
                        U[i][i] = A2[i][i]       # Making diagonal elements of U equal to A[i][i] 
                if j>i:
                    U[i][j] = A2[i][j]    # Separating L and U from the L/U augmented form obtained from crout()
                            
        y = forsubs(L,b)        # Solving Ly=b
        x = backsubs(U,y)       # Finally solving Ux=y
        x = roundMat(x,3)       
        return(x)
    else:
        return("No unique soln exists/LU Decomposition failed")
        exit()

# Inverse of a matrix using LU (Doolittle algorithm) decomposition
def invMat(M):
    n = len(M)
    I = createI(n)
    A = doolittle(M)  # Performing LU decomposition on A using doolittle()
    
    if A=="LU Decomposition not possible":
        return("Inverse doesn't exist/LU decomposition not possible for the given matrix.")
    
    elif detMat(M)==0:
        return("Det = 0. Inverse does not exist.")
    
    else:
        # Separating L and U
        if A!=None:          # Decomposing A2 into L and U using Doolittle
            L = create0(n,n)
            U = create0(n,n)
            for i in range(n):
                for j in range(n):
                    if j<i:
                        L[i][j] = A[i][j]
                    if j>=i:
                        U[i][j] = A[i][j]    # Separating L and U from the L/U augmented form obtained from doolittle()
                        if j==i:
                            L[i][i] = 1       # Making diagonal elements of L to 1

            # We now have L and U separately stored, which can be used.
            inv = create0(n,n)
            for i in range(n):
                temp1 = extractCol(I,i+1)  # Exctracting columns of I one by one for solving.
                y = forsubs(L,temp1)      # Solving Ly = I[i] (ith column of I)
                x = backsubs(U,y)         # Solving Ux = A^(-1)[i] (ith column of A^(-1))
                x = roundMat(x,2)
                for j in range(n):
                    inv[i][j]=x[j][0]
            return inv
        else:
            return("No unique soln exists/LU Decomposition failed")
            exit()
```

## Complex Numbers

```python
# Modulus
def modComplex(r,i):
    mod = (r**2+i**2)**(1/2)
    return mod
    
# Conjugate
def conjugate(r,i):
    i = -i
    return r,i
    
# Addition
def addComplex(r1,i1,r2,i2):
    sum_r = r1+r2
    sum_i = i1+i2
    return sum_r, sum_i
    
# Product
def prodComplex(r1,i1,r2,i2):
    prod_r = r1*r2 - i1*i2
    prod_i = i1*r2 + r1*i2
    return prod_r, prod_i
    
# Division
def divComplex(r1,i1,r2,i2):
    div_r = r1*r2 + i1*i2
    div_i = i1*r2 - r1*i2
    mod = (r2**2+i2**2)
    div_r = div_r/mod
    div_i = div_i/mod
    return div_r, div_i

# Phase angle
def phaseComplex(r,i):
    import math
    theta=math.atan(abs(i/r))
    if r>0 and i>0:
        return theta
    elif r<0 and i>0:
        return (math.pi-theta)
    elif r>0 and i<0:
        return (-theta)
    elif r<0 and i<0:
        return (theta-math.pi)

        
# For printing a complex number in proper format
def printComplex(r,i):
    if i>=0:
        print(str(r)+" + "+str(i)+"i")
    else:
        print(str(r)+" - "+str(-i)+"i") 
```

```python
## BRACKETING ##
def bracket(func,low,high,maxm=12):
    # maxm stores the maximum no. of iterations allowed.
    truth = func(low)*func(high)<=0
    i = 0
    if truth == False:
        while i <maxm and truth == False: 
            if abs(func(low)) < abs(func(high)):
                low = low - 0.5*(high-low)
                i = i+1
            elif abs(func(high)) < abs(func(low)):
                high = high + 0.5*(high-low)
                i = i+1
            truth = func(low)*func(high)<=0        
        if i ==12:        
            return "bad bracket try some other interval"
        else:
            return [low,high]  
    else:
        return [low,high]         

## BISECTION ## 
# Obtaining roots of a function using Bisection
def bisection(func,a,b,err= 10**(-7)):
    j=1
    if func(a)*func(b)>0:
        print("Do proper bracketing")
    else:
        while (b-a)/2 > err:
            c = (a+b)/2
            if (func(c) == 0):
                break
            if (func(c)*func(a)<0):
                b =c
            else:
                a=c  
            j = j+1
        return c

# A tailored form of the above function designed for As5-Q1
# Will turn this into a multi-line comment later.
def bisection_as1(func,a,b,err= 10**(-4)):
    plot1 =[]  # Stores the values of successive f(x_i)
    plot2 =[]  # Stores the values of j (# of iterations)
    plot3 =[]  # Stores the succesive x_i
    j=1
    if func(a)*func(b)>0:
        print("Do proper bracketing")
    else:
        while (b-a)/2 > err:
            c = (a+b)/2
            if (func(c) == 0):
                break
            if (func(c)*func(a)<0):
                b =c
            else:
                a=c  
            j = j+1
            plot1.append(func(c))
            plot2.append(j)
            plot3.append(c)
        return c, plot1, plot2, plot3

# A Function which returns the roots with tabulation of
# convergence as well as the convergence plot
def bisection_with_convergence(f,a,b):
    r,m,n,o = bisection_as1(f,a,b)
    
    '''
    - r stores the returned root obtained using bisection.
    - m stores the values of f(x_i)
    - n stores the values of i
    - o stores the values of x_i
    '''

    # Tabulating the convergence
    print("Table for convergence:")
    print("i       x_i        f(x_i)")
    print("-------------------------")
    for i in range(len(m)):
        print(f"{n[i]}      {round(o[i],4)}      {round(m[i],4)}") 
        # Using round() to print only upto 4 decimals just to make the output neat.
    print()


    # Plotting the convergence
    from matplotlib import pyplot as plt
    plt.plot(n,m)
    plt.xlabel(" No. of iterations (i) ")
    plt.ylabel("f (x_i)")
    plt.title("Convergence Plot")
    plt.show()

    # Printing the root
    print()
    print("---------------------------------------------")
    print(f"Root in the given interval obtained using:")
    print(f"    o Bisection method: {r}")   # Result using Bisection method
```

### Laguerre's method

```python
## Synthetic Div ##
def synDiv(coeff,r):
    ans=[]   # List for storing coefficients of Q(x)
    temp=[]  # List for creating the matrix/addend to be added
    temp.append(0)          # Assigning the 1st place of the addend as 0
    for i in range(len(coeff)-1):      # Iterating through each coefficient
        ans.append(coeff[i]+temp[i])        
        temp.append(r*ans[i])   # Note that the iteration takes place only upto the
                                # second last place,but the last value has been assigned to temp
    #ans.append(coeff[-1]+temp[-1])        # Assigning the last place, i.e. the remainder
    #if ans[-1]!=0:
    #    return("Bad Bracket.")
    #else:
        #ans.pop(-1)        # Removing the last coefficient as it is 0
    return ans

# Function for differentiating (only for 1st or second order differentiation)
def differentiator(f,n,x,h=10**(-5)):    # Numerical nth (n = 1 or 2) Differentiation of f(x)
    if int(n)==1:                        # at x; h is the tolerance
        df = round((f(x+h)-f(x))/h,2)
        return df
    elif int(n)==2:
        d2f = round((f(x+h)+f(x-h)-2*f(x))/(h**2),2)
        return d2f

# Function for returning root by taking a guess as input
def calcLag(coeff,deg,alpha,err=10**(-5)):
    # coeff = list of coefficients, deg = degree of poly., alpha = guess, err = tolerance
    # Importing math for calculating the square root
    import math
    # writing the function from the list of coefficients
    f=lambda x: sum([coeff[i]*x**(deg-i) for i in range(deg+1)])
    roots=[]
    while True:  # Running an infinite loop which breaks iff alpha is a root
        if abs(f(alpha))<err:          # Checking if alpha is a root
            ### NEWTON-RAPHSON (for polishing of roots) ###
            alpha = newtonRaphson(f,alpha)
            ans = synDiv(coeff,alpha)  # Performing synthetic division
            return alpha,ans
            break
        else:      # ROOT-FIXING (when alpha is not a root)
            df = differentiator(f,1,alpha)   # Numerically taking the 1st derivative of f at alpha
            d2f = differentiator(f,2,alpha)  # Numerically taking the 2nd derivative of f at alpha
            # Note that df and d2f are not functions, but numerical values.
            G = df/f(alpha)
            H = G**2 - d2f/f(alpha)    # Also note that G and H are numerical values, not functions.
            # sq includes the term in square root such that it gives maximum value to the denominator.
            sq = max(G+math.sqrt((deg-1)*(deg*H-G**2)),G-math.sqrt((deg-1)*(deg*H-G**2)))
            alpha=alpha-deg/sq
            
# Driver code for Laguerre's method
def laguerre(coeff,alpha,err=10**(-5)):
    roots=[]             # This list will store the roots
    deg=len(coeff)-1     # Storing the degree of entered function
    while deg>1:     # Deflating till 2nd degree
        alpha, coeff = calcLag(coeff,deg,alpha) # alpha,coeff store the new alpha and coeff (after synDiv)
        roots.append(alpha)    # Appending the newly obtained root to the list
        deg = deg-1
    # when deg=1, deflation isn't necessary; so, we only work out the algebra
    # for deg = 1, coeff is in the form [c1,c2]
    # coeff[0]*x + coeff[1] = 0 ; => x = -coeff[1]/coeff[0]
    # roots.append(round(-coeff[1]/coeff[0],3))
    roots.append(-coeff[1]/coeff[0])
    return (roots)
```

### Regula-Falsi method

```python
# Finding roots using Regula-Falsi / Falsification method
def regulaFalsi(f,a,b,err=10**(-6)):
# f = the function; [a,b] = The interval to check.
    c=b-(b-a)*f(b)/(f(b)-f(a))
    while True:       # Running an ifinite loop till a root is found.
        temp=c
        c=b-(b-a)*f(b)/(f(b)-f(a))    
        if f(a)*f(c)<0:
            b=c
            a=a
        elif f(a)*f(c)>0:
            b=b
            a=c
        if abs(c-temp)<err and abs(f(c))<err:
            return(c)
```

### Newton-Raphson method

```python
#Solving Equation using Newton-Raphson method
def newtonRaphson(f,x_0,err=10**(-5)):
# f = The function, x_0 = Initial guess
    n=1
    if abs(f(x_0))==0:
        return(x_0)
    x=x_0-f(x_0)/differentiator(f,1,x_0,err*10**(-3))
    while True:
        n+=1
        x_bef=x
        x=x-f(x)/differentiator(f,1,x)
        if abs(x-x_bef)<err and abs(f(x))<err:
            return(x)
```

# NUMERICAL INTEGRATION

```python
''' MIDPOINT METHOD '''
#f = function to be integrated; [a,b] = Interval of integration; 
#N = No. of steps
def midpoint_int(f,a,b,N):
    h = float(b-a)/N
    value = 0
    for i in range(N): 
        value = value + f((a + h/2.0) + i*h)
    return h*value

''' TRAPEZOIDAL METHOD '''
def trapezoidal_int(f,a,b,N):
    h = float(b-a)/N
    ans = 0
    for i in range(N):
        ans = ans + f(a +i*h) + f(a +(i+1)*h)
    return h*ans/2

''' SIMPSON METHOD '''
def simpson_int(f,a,b,N):
    h = float(b-a)/N
    ans = f(a) + f(b)
    for i in range(N):
        x = a + i*h
        if i%2 == 0:
            ans = ans + 2*f(x)
        else:
            ans = ans + 4*f(x)
    return ans*h/3

''' MONTE CARLO INTEGRATION '''
def montecarlo_int(f,a,b,N):
    import random
    integral = 0
    val =0
    for i in range(N):
        x = a + (b-a)*random.random()
        val += f(x)
    integral = float(((b-a)*val)/N)    
    return integral

''' MONTE CARLO with plot '''
# Monte-Carlo integration followed by the plotting 
# of integral value and standard dev.
def montecarlo_with_plot(f,a,b,N):
    import random
    from matplotlib import pyplot as plt
    value3 = []
    iter = []
    devi = []
   
    for j in range(300):
        integral = 0
        val =0  # integral
        val1 = 0     # standard deviation
        for i in range(N):
            x = a + (b-a)*random.random()
            val += f(x)
            val1 += (f(x))**2  
                                       
        integral = float(((b-a)*val)/N) 
        sd =  sqrt((1/N)*(val1)- ((1/N)*(val))**2) 
        devi.append(sd)     
        value3.append(integral)
        iter.append(N)
        N += 10
    # NOTE that N is a random number which is set to increase in multiples of 10.
    print(f"Value of the integral calculated using Monte Carlo Integration = {montecarlo_int(f,a,b,N)}")
    a,b,c = value3,iter,devi
    
    plt.figure(0)
    plt.plot(b,a,"b-")   # plotting the integral value
    plt.xlabel("N")
    plt.ylabel("Integral value")

    plt.figure(1)
    plt.plot(b,c,"g-") # plotting the standard deviation
    plt.xlabel("N")
    plt.ylabel("Std. Deviation")
```
