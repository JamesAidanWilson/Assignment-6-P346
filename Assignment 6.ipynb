```python
''' QUESTION 1 '''
import math
from tabulate import tabulate   # Importing tabulate to ctreate neat tables
f = lambda x: math.sqrt(1+(1/x))
                        # Using MIDPOINT method
m8 = midpoint_int(f,1,4,8)     # N = 8
m16 = midpoint_int(f,1,4,16)   # N = 16
m24 = midpoint_int(f,1,4,24)   # N = 24
                        # Using Trapezoidal method
t8 = trapezoidal_int(f,1,4,8)    # N = 8
t16 = trapezoidal_int(f,1,4,16)  # N = 16
t24 = trapezoidal_int(f,1,4,24)  # N = 24
                        # Using Simpson method
s8 = simpson_int(f,1,4,8)      # N = 8
s16 = simpson_int(f,1,4,16)    # N = 16
s24 = simpson_int(f,1,4,24)    # N = 24
# The analytical value
ref = 3.62
print("Analytical value of the integral = 3.62")
print()
err=[]
data=[m8,m16,m24,t8,t16,t24,s8,s16,s24]
for x in data:
    err.append(abs(ref-x))
# Tabulating
print(tabulate([['Midpoint',m8,err[0],m16,err[1],m24,err[2]],['Trapezoidal',t8,err[3],t16,err[4],t24,err[5]],['Simpson',s8,err[6],s16,err[7],s24,err[8]]],headers=['Integration Method','N = 8','Error','N = 16','Error','N = 24','Error']))
```
